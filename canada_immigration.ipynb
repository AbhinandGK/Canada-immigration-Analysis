## Scatter Plots

```python
df_total=pd.DataFrame(df[years].sum())
df_total.index=map(int, df_total.index)
df_total.reset_index(inplace = True)
df_total.columns=['Year','Total']
df_total.head()
```

```python
df_total.plot(kind='scatter',x='Year',y='Total',figsize=(20,10))
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
```

```python
x=df_total['Year']
y=df_total['Total']
fit=np.polyfit(x,y,deg=1)

df_total.plot(kind='scatter',x='Year',y='Total',figsize=(20,10))
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.plot(x,fit[0]*x+fit[1],color='red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2005, 150000))
```

Here we get the equation,
Number of Immigrants=5567xYear - 10926195

Using the equation of line of best fit, we can estimate the number of immigrants in 2019

No of Immigrants in 2019 =5567*2019-10926195
                         =313578



According to wikipedia the total immigrants to canada in 2019 is 341,180. Our estimated value of 313578 is within 8.1% of the actual number
