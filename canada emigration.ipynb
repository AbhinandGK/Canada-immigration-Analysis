# Canada Data Analysis

import importent libraries

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Read xl files

```python
df=pd.read_excel('Canada.xlsx',sheet_name='Canada by Citizenship',skiprows=range(20),skipfooter=2)
```

```python
df.head()
```

## Drop unwanted data 

```python
df.columns.values
```

```python
df.drop(['DEV','Coverage','AREA','REG','Type'], axis=1 ,inplace=True)
```

```python
df.head()
```

## Rename Column names

```python
df.rename(columns={'AreaName' : 'Continent', 'RegName':'Region', 'OdName':'Country', 'DevName':'Dev status'},inplace=True)
df.head()
           
```

```python
df['Total']=df.sum(axis=1)
```

```python
df.head()
```

## Basic Overview of  the Data

```python
df.describe()
```

## Set index as Country name

```python
df.set_index('Country',inplace=True)
df.head()
```

```python
df.loc['India']
```

```python
df.columns=list(map(str, df.columns))
years=list(map(str,range(1980,2014)))
```

```python
condition=df.Continent=='Asia'
df[condition]
```

```python
%matplotlib inline
```

```python
print(plt.style.available)
```

```python
india=df.loc['India',years]
india.head()
```

```python
india.plot(kind='barh',figsize=(15,15))
plt.title('Immigration from India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

here we can see immegration more in 2005. After google search we get that immigration was high primarily due to settlement in the Western provinces

```python
data=df.loc[['India','China'],years]
data=data.transpose()
data.plot(kind='line')
plt.xlabel('year')
plt.ylabel('Immegration')
plt.title('India vs China Immigration')
```

## Top contries that contributed to most immigrants

```python
df.sort_values(['Total'],inplace=True,axis=0,ascending=False)
```

```python
top=df.head(5)
top=top[years].transpose()
```

```python
top.plot(kind='area',figsize=(20, 10),stacked=True)
plt.title('Immigration Trend of Top 5 Countries(Stacked)')
plt.xlabel('Number of Immigrants')
plt.ylabel('Years')
```

```python
top.plot(kind='area',figsize=(20, 10),stacked=False,alpha=.3)
plt.title('Immigration Trend of Top 5 Countries')
plt.xlabel('Number of Immigrants')
plt.ylabel('Years')
```
